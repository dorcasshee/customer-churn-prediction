# API Query Example - Customer Churn Prediction

```python
import requests
website = "http://localhost:5000"
```

### 1. Make a Prediction via HTTP POST API

```python
url = f'{website}/api/predict_one'
model_inputs = {   
    'age': 45,
    'senior_citizen': 'no',
    'married': 'yes',
    'num_dependents': 1,
    'tenure_months': 32,
    'num_referrals': 2,
    'has_internet_service': 'yes',
    'has_unlimited_data': 'yes', 
    'has_premium_tech_support': 'no',
    'paperless_billing': 'yes',
    'total_monthly_fee': 65, 
    'total_charges_quarter': 2280,
    'avg_long_distance_fee_monthly': 23,
    'avg_gb_download_monthly': 20
}

response = requests.post(url, json = model_inputs)
print(response)  # 200 means OK, 400 means Bad Request
print(response.text)
```

### 2. Multiple Predictions using one API call

```python
import pandas as pd
file_id = '16or9tsk2n91qGXstOOtlHCIlhkkE_NZo'
url = f'https://drive.google.com/uc?id={file_id}'

data = pd.read_csv(url).drop("Unnamed: 0", axis=1)
data.head()
```

```python
records = data.to_dict('records')
records[:2]
```

```python
url = f'{website}/api/predict'

response = requests.post(url, json = records)
print(response)             # 200 means OK, 400 means Bad Request
print(response.text[:150])  # print first 150 characters of response
```
